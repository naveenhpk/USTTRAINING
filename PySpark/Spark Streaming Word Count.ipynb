```python
from pyspark.streaming import StreamingContext
from pyspark.streaming.kafka import KafkaUtils
from pyspark.sql import SparkSession 
from pyspark import SparkConf, SparkContext
```

### 1.CREATE SPARK STREAMING CONTEXT

```python
config =SparkConf().setAppName('SparkSession').setMaster('local[4]')
sc = SparkContext.getOrCreate(conf=config)
```

```python
#initiialize streaming context with batch interval of five seconds
ssc = StreamingContext(sc,5)
```

```python
#dstream format tht will connect to url with localhost as IP AND Port number 9990
lines = ssc.socketTextStream("localhost",9990)
words = lines.flatMap(lambda line : line.split(" "))
pairs = words.map(lambda word : (word,1))
wordcount = pairs.reduceByKey(lambda x,y:x+y)
wordcount.pprint()
```

```python
ssc.start()
ssc.awaitTermination()
```

```python
ssc.stop()
```

```python
sc.stop()
```
